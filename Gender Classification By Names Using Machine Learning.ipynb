IMPORTING SOME LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

Load the Dataset

```python
df = pd.read_csv('/content/Names_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['gender'] = df['gender'].replace({'f':0,
                                     'm':1})
```

```python
name=df['name']
```

Apply CountVectorizer

```python
cv = CountVectorizer()
```

```python
x = cv.fit_transform(name)
```

```python
from scipy.sparse import csr_matrix
cv.get_feature_names_out()
```

```python
#here x = cv.fit_transform(name)
y=df['gender']
```

Split the data into train data and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x.shape,x_train.shape,x_test.shape
```

Apply MultinomialNB

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.predict(x_test)
```

```python
test_accuracy = mnb.score(x_test,y_test)
print('test_accuracy {} %'.format(test_accuracy*100))
```

Building a prediction system for evaluating

```python
input = ['Amelia']
cv = cv.transform(input)
prediction = mnb.predict(cv)

if (prediction==0):
    print('Female')
else:
    print('Male')
```
